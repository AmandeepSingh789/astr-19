```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x=[]
for z in range(1,101):
    y= z/100
    x.append(y)
```

```python
def return_exp(input):
    return(np.exp(input))
```

```python
y= return_exp(x)
plt.plot(x,y)
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')

plt.savefig('S7.pdf', dpi=120, format='pdf', bbox_inches='tight')
plt.show()
```
