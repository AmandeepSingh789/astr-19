```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x=[]
for z in range(1,101):
    y= z/100
    x.append(y)
```

```python
def returnSin(input):
    return(np.sin(input))
```

```python
def returnCos(input):
    return(np.cos(input))
```

```python
a =returnSin(x)
b =returnCos(x)
f, axarr=plt.subplots(1,2)

axarr[0].plot(x,a)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')


axarr[1].plot(x,b)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')
f.subplots_adjust(wspace=0.4)


plt.savefig('S8.pdf', dpi=120, format='pdf', bbox_inches='tight')
plt.show()
```
